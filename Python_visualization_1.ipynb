```python
from math import exp
x=list(range(0,20))
y=[exp(i) for i in x]
plt.plot(x,y)
```

```python
#plot multiple plot
from math import *
plt.figure(figsize=(16,9))
x=list(range(-10,10))
y=[i*i for i in x]
plt.subplot(3,4,1)  #3X4 grid,1st position of graph

plt.plot(x,y)
plt.title("graph of squares")

z=[i*i*i for i in x]
plt.subplot(3,4,6) #8th position
plt.plot(x,z)
plt.title("Graph of cubes")

w=[sin(i) for i in x]
plt.subplot(3,4,12)#12th position
plt.plot(x,w)
plt.title("Graph of sin(x)")



```
