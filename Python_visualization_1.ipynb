```python
#plot multiple plot
from math import *
plt.figure(figsize=(16,9))
x=list(range(-10,10))
y=[i*i for i in x]
plt.subplot(3,4,1)  #3X4 grid,1st position of graph

plt.plot(x,y)
plt.title("graph of squares")

z=[i*i*i for i in x]
plt.subplot(3,4,6) #8th position
plt.plot(x,z)
plt.title("Graph of cubes")

w=[sin(i) for i in x]
plt.subplot(3,4,12)#12th position
plt.plot(x,w)
plt.title("Graph of sin(x)")



```

```python
#swarm plot
sns.swarmplot(x="day",y="tip",data=tips,size=6)
plt.ylabel("tip($)")
```

```python
sns.swarmplot(x="day",y="tip",data=tips,size=6,hue="sex")
plt.ylabel("tip($)")
plt.show()
```

```python
sns.swarmplot(x="day",y="tip",data=tips,size=6,hue="time")
plt.ylabel("tip($)")
plt.show()
```

```python
#subplot
plt.subplot(2,2,1)
sns.boxplot(x="day",y="tip",data=tips)

plt.subplot(2,2,2)
sns.barplot(x="day",y="tip",data=tips)
plt.ylabel("tips($)")
plt.title("boxplot of tips")


plt.subplot(2,2,3)
sns.violinplot(x="day",y="tip",data=tips)

plt.subplot(2,2,4)
sns.swarmplot(x="day",y="tip",data=tips)
plt.ylabel("tips($)")
plt.title("violint of tips")
plt.tight_layout()
plt.show()
```

```python
figbi, axesbi = plt.subplots(2,4, figsize=(18,10))
data.groupby("Pclass")["Survived"].mean().plot(kind="bar",ax=axesbi[0,0],xlim=[0,1])
data.groupby("SibSp")["Survived"].mean().plot(kind="bar",ax=axesbi[0,1],xlim=[0,1])
data.groupby("Parch")["Survived"].mean().plot(kind="bar",ax=axesbi[0,2],xlim=[0,1])
data.groupby("Sex")["Survived"].mean().plot(kind="bar",ax=axesbi[0,3],xlim=[0,1])
data.groupby("Embarked")["Survived"].mean().plot(kind="bar",ax=axesbi[1,0],xlim=[0,1])
sns.boxplot(x="Survived",y="Age",data=data, ax=axesbi[1,1])
sns.boxplot(x="Survived",y="Fare",data=data, ax=axesbi[1,2])
plt.show()
figbi.savefig("myfig.png") #or ("myfig.jpeg")
from google.colab import files
files.download("myfig.png")


```
